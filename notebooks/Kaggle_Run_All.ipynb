# Kaggle Run-All: DQN CartPole
This notebook runs pretraining, RL training, evaluation, and chart generation end-to-end.


```python
import os, sys, subprocess
print('Python:', sys.version)
print('CWD:', os.getcwd())
```

```python
commands = [
    'python pretrain.py',
    'python train.py --episodes 500 --pretrained checkpoints/pretrained_model.pkl',
    'python evaluate.py --model checkpoints/best_model.pkl',
    'python scripts/generate_all_charts.py'
]
for c in commands:
    print('\n>>>', c)
    rc = subprocess.call(c, shell=True)
    print('return code:', rc)
    if rc != 0:
        raise SystemExit(f'Command failed: {c}')
```

```python
import json
with open('logs/training_stats.json') as f:
    stats = json.load(f)
print('Best reward:', stats.get('best_reward'))
print('Total steps:', stats.get('total_steps'))
print('Solved episode:', stats.get('solve_episode'))
```
